```python
import os
from pathlib import Path
from sunpy.net import Fido
from sunpy.net.attrs import Instrument, Level, Time
from sunpy_soar.attrs import Product

# --- Create attributes for the query ---
instrument = Instrument('METIS')

# Time interval
time = Time('2022-09-15 2:00:00', '2022-09-15 23:00:01')

# Data Level (L2 in this case)
level = Level(2)

""" Download pB images """
# download_path = f'{str(Path().resolve().parent)}/data/input/VL_IMAGES/'
# product_descriptor = Product('metis-vl-pb')

""" Download UV images """
download_path = f'{str(Path().resolve().parent)}/data/input/UV_IMAGES/'
product_descriptor = Product('metis-uv-image')

os.makedirs(download_path, exist_ok=True)
print(f"Files will be downloaded in: {os.path.abspath(download_path)}")




# --- Pass all the attributes to Fido for the search ---
result = Fido.search(instrument, time, level, product_descriptor)

# Stampa i risultati della ricerca
print("\nQuery results:")
print(result)

# --- Download ---
if result:
    print("\nDownloading...")
    files = Fido.fetch(result, path=download_path)
    print("\nFile scaricati:")
    print(files)
else:
    print("\nNo file matched the specified attributes.")

```
